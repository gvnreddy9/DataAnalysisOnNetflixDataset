```python
query = f"""
WITH cte AS (
  SELECT *,
  ROW_NUMBER() OVER (PARTITION BY title, type ORDER BY show_id) AS rnk
  FROM project.netflix_raw
) 
select * from cte where rnk =1 order by show_id;
"""
resoverall = engine.execute(query)
dfN = pd.DataFrame(resoverall.fetchall())
dfN.columns = resoverall.keys()
```

```python
def split_director(director_string):
    if director_string is None:
        return []  # Return an empty list for None values
    else:
        return director_string.split(",")
```

```python
director_df = pd.DataFrame(
    {
        "show_id": dfN["show_id"],
        "director": dfN["director"].apply(split_director).explode(),
    }
)

country_df = pd.DataFrame(
    {
        "show_id": dfN["show_id"],
        "country": dfN["country"].apply(split_director).explode(),
    }
)
genre_df = pd.DataFrame(
    {
        "show_id": dfN["show_id"],
        "genre": dfN["listed_in"].apply(split_director).explode(),
    }
)
cast_df = pd.DataFrame(
    {
        "show_id": dfN["show_id"],
        "cast": dfN["cast"].apply(split_director).explode(),
    }
)
```

```python
director_df.dropna(subset=['director'],inplace=True)
country_df.dropna(subset=['country'],inplace=True)
genre_df.dropna(subset=['genre'],inplace=True)
cast_df.dropna(subset=['cast'],inplace=True)
```

```python

#dfNew = df.rename(columns={'B': 'Beta'})

director_df.to_sql(
    "netflix_directors",
    engine,
    if_exists="append",  # Specify how to handle existing table
    index=False,  # Don't include the index column
)

country_df.to_sql(
    "netflix_country",
    engine,
    if_exists="append",  # Specify how to handle existing table
    index=False,  # Don't include the index column
)
genre_df.to_sql(
    "netflix_genre",
    engine,
    if_exists="append",  # Specify how to handle existing table
    index=False,  # Don't include the index column
)

cast_df.to_sql(
    "netflix_cast",
    engine,
    if_exists="append",  # Specify how to handle existing table
    index=False,  # Don't include the index column
)
```

```python
query = f"""
WITH cte AS (
  SELECT *,
  ROW_NUMBER() OVER (PARTITION BY title, type ORDER BY show_id) AS rnk
  FROM project.netflix_raw
) 
select show_id,type,title,cast(date_added as date) as date_added,release_year
,rating,case when duration is null then rating else duration end as duration,description 
from cte;
"""
resoverall = engine.execute(query)
dfNN = pd.DataFrame(resoverall.fetchall())
#dfNN.columns = resoverall.keys()
```

```python
dfNN.sort_values(by=['show_id'],inplace=True)
```
